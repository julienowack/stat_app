# Manipulations des bases de données
Pour toutes les bases, il s'agit de les réduire à Paris et éventuellement de les grouper par mois

Bases de données du traffic aérien

```python
df_air_traffic=pd.read_csv('df_air_traffic.csv')
df_air_traffic_paris=df_air_traffic[df_air_traffic['citydest']=='Paris']
df_air_traffic_paris['sheduled_time_start']=pd.to_datetime(df_air_traffic_paris['sheduled_time_start'])
df_at_paris_month = df_air_traffic_paris.groupby(pd.Grouper(key='sheduled_time_start', freq="MS")).sum()
df_at_paris_month=df_at_paris_month.sort_values('sheduled_time_start')
df_at_paris_month=df_at_paris_month.reset_index()
df_at_paris_month=df_at_paris_month[['sheduled_time_start', 'seats', 'number_planes']]
```

Base de données des Google Trends

```python
df_gt_paris= pd.read_csv('df_gt_paris.csv')
df_gt_paris['date']=pd.to_datetime(df_gt_paris['date'])
df_gt_paris_month = df_gt_paris.groupby(pd.Grouper(key='date', freq="MS")).sum()
df_gt_paris_month=df_gt_paris_month.sort_values('date')
df_gt_paris_month=df_gt_paris_month.reset_index()
df_gt_paris_month
```

Base de données des commentaires utilisateurs

```python
df_reviews_us_france_210712 = pd.read_csv('reviews_us_france_210712.csv')
df_reviews_france=df_reviews_us_france_210712[df_reviews_us_france_210712['country']=="France"]
df_reviews_paris=df_reviews_france[df_reviews_france['city']=='Paris']
df_reviews_paris['stay_date']=pd.to_datetime(df_reviews_paris['stay_date'])
df_reviews_paris_month = df_reviews_france.groupby("stay_date", as_index=False).sum()
df_reviews_paris_month
```
